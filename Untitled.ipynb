```python
samples = []
for _ in range(sample_size):
    random_length = np.random.rand() * total_length
    temp_total_length = random_length
    for i, segment in enumerate(segments):
        if temp_total_length - segment < 0:
            break
        temp_total_length = temp_total_length - segment

    left_vertex = simplices[i][0]
    right_vertex = simplices[i][1]

    sample = vertice[left_vertex] + (vertice[right_vertex] - vertice[left_vertex]) * temp_total_length / segments[i]
    samples.append(sample)
samples = np.array(samples)

mean_samples = np.average(samples, axis=0)
Y = [np.dot((sample - mean_samples).reshape(-1,1), (sample - mean_samples).reshape(-1,1).T) for sample in samples]
print(np.average(Y, axis=0))
Y = np.power(np.average(Y, axis=0), -1/2)
transformed_vertice = np.dot(Y, vertice.T).T
```

```python
import itertools

def metrize(mat):
    
    def check_square_rank(mat):
        return mat.shape[0] == mat.shape[1]
        
    def positivate(mat):
        for i in range(1, mat.shape[0]):
            for j in range(0, i-1):
                mat[j,i] = np.abs(mat[j,i])
                
    def parallelize(mat):
        for i in range(1, mat.shape[0]):
            for j in range(0, i-1):
                mat[i,j] = mat[j,i]
    
    def list_all_triangles(mat):
        elements = range(mat.shape[0])
        return list(itertools.combinations(elements, 3))
                
    def trianglize(mat):
        triangles = list_all_triangles(mat)
        
        def triangle_to_lengths(triangle):
            lengths = []
            for i in range(3):
                j = i+1 if i != 2 else 0
                    
                lengths.append(mat[triangle[i],triangle[j]])
                
            return np.array(lengths)
        
        for triangle in triangles:
            lengths = triangle_to_lengths(triangle)
            max_ind = np.argmax(lengths)
            max_value = lengths[max_ind]
            sum_other_values = np.sum(lengths[np.array([0,1,2] != max_ind)])
            if max_value > sum_other_values:
                j = max_ind +1 if max_ind != 2 else 0
                mat[triangle[j], triangle[max_ind]] = sum_other_values
                
        parallelize(mat)
    
    if not check_square_rank(mat):
        raise
        
    positivate(mat)
    trianglize(mat)
```
